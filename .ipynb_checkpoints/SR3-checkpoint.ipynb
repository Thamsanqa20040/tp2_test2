```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sihle =pd.read_csv("C:\\Users\\admin\\Downloads\\Informal_Settlement_Programme_-1898186875815614491.csv")
```

```python
print(sihle.dtypes)
```

```python
print(sihle.columns)
```

```python
subset_sihle = sihle.head(12) 

plt.figure(figsize=(12, 6))
sns.barplot(x='STREETNAME', y='EST_HHC',  data=subset_sihle)
plt.title('The estimated household per stree name')
plt.xlabel('The street names')
plt.ylabel('Estimated household count per street')
plt.xticks(rotation=45)
plt.show()
```

```python
subset_sihle = sihle.head(12)

plt.figure(figsize=(12, 6))
plt.scatter(subset_sihle['Shape__Area'], subset_sihle['EST_HHC'])
plt.title('The shape area of the household count')
plt.xlabel('The saphe area of the street')
plt.ylabel('Estimated houshold per street')
plt.xticks(rotation=45)
plt.show()
```

```python
subset_sihle = sihle.head(12)

plt.figure(figsize=(12, 6))
plt.pie(subset_sihle['EST_HHC'], labels=subset_sihle['STREETNAME'], autopct='%1.1f%%', startangle=140)
plt.title('Total Water needed to be supplied to the  per Street')
plt.show()
```

```python
subset_sihle = sihle.head(12)

# Plotting the box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='EST_HHC', data=subset_sihle)  # Corrected the column name (remove extra space)
plt.title('The Estimated Household')
plt.xlabel('Street Name')
plt.ylabel('Estimated Household Count')  # Added a label for the y-axis
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the histogram
plt.figure(figsize=(8, 6))  # Changed the figure size to (8, 6)
sns.histplot(data=sihle['OBJECTID'], bins=5, kde=True)  # Assuming 'sihle' is your DataFrame
plt.title('Distribution of OBJECTID')
plt.xlabel('OBJECTID')
plt.ylabel('Frequency')
plt.show()
```
